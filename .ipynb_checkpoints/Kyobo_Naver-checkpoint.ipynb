```python
import pandas as pd
import numpy as np

import platform
import matplotlib.pyplot as plt

%matplotlib inline

path = "C:\Windows\Fonts\malgun.ttf"
from matplotlib import font_manager, rc
if platform.system() == 'DarWin':
    rc('font', family = 'AppleGothic')
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family = font_name)
else:
    print ('Unknown system... sorry~!')
    
plt.rcParams['axes.unicode_minus'] = False

from bs4 import BeautifulSoup
from urllib.request import urlopen
import urllib
import time


tmp1 = 'https://search.naver.com/search.naver?where=kin'
html = tmp1 + '&sm=tab_jum&ie=utf8&query={key_word}&start={num}'

response = urlopen(html.format(num=1, key_word=urllib.parse.quote('교보생명 추천상품')))
soup = BeautifulSoup(response, "html.parser")
tmp = soup.find_all('dd')
tmp
```

```python
tmp_list = []
for line in tmp:
    tmp_list.append(line.text)
    line.text
tmp_list
```

```python
from tqdm import tqdm_notebook

present_candi_text = []

for n in tqdm_notebook(range(1, 3000, 10)):
    response = urlopen(html.format(num=n, key_word=urllib.parse.quote('교보생명 추천상품')))
    
soup = BeautifulSoup(response, "html.parser")

tmp = soup.find_all('dd')

for line in tmp:
    present_candi_text.append(line.text)
    
time.sleep(30)
```

```python
import nltk
from konlpy.tag import Twitter; t=Twitter()

present_candi_text
```

```python
present_text = ''
for each_line in present_candi_text:
    present_text = present_text + each_line + '\n'
    #present_text = present_text.replace('금값','금')
#tokens_ko = t.morphs(present_text)
tokens_ko = t.nouns(present_text)
tokens_ko
```

```python
stop_words = ['추천','년','카페','전','가입','만원','무','금액','리','분','증권','약','이상','월','수익','게','첫','갯수','대략',
              '아시','중','이','좀','얼마','하라','개','질문','삼성','한화','흥국생명'
             '수','만','제','때','안','보','알','문서','문장','반드시','쉼표','여러','세','억','표','률','저','다른','시분',
             '새로이','정도','사분','주','드릴','네이버','상','직접','막','보통','과','제대로','사은','차일','답변','닉네임']
tokens_ko = [each_word for each_word in tokens_ko if each_word not in stop_words]

ko = nltk.Text(tokens_ko, name='교보생명 추천상품')
print(len(ko.tokens))
print(len(set(ko.tokens)))

ko.vocab().most_common()
```

```python
plt.figure(figsize=(15,6))
ko.plot(50)
plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
```

```python
data = ko.vocab().most_common(800)

wordcloud = WordCloud(font_path='C:\Windows/Fonts/malgun.ttf',
                     relative_scaling = 0.5,
                     stopwords = STOPWORDS,
                     background_color = 'white',
                     ).generate_from_frequencies(dict(data))
plt.figure(figsize=(16,8))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
