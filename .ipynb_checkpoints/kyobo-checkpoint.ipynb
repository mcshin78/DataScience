```python
import nltk

from konlpy.corpus import kobill
files_ko = kobill.fileids()
doc_ko = kobill.open('kyobo.txt').read()
doc_ko
```

```python
from konlpy.tag import Twitter; t=Twitter()
tokens_ko = t.nouns(doc_ko)
tokens_ko
```

```python
stop_words = ['등', '및', '수', '모든', '임', '제', '대한', '장', '창', '와', '바', '중', '경우', '것', '그', '관련', '후', '내', '때']
ko = [each_word for each_word in tokens_ko if each_word not in stop_words]
```

```python
ko = nltk.Text(ko, name='교보생명 직무윤리 실천규범')
```

```python
import matplotlib.pyplot as plt
import platform

path = "C:\Windows\Fonts\malgun.ttf"
from matplotlib import font_manager, rc
if platform.system() == 'Darwin' :
    rc('font', family="AppleGothic")
elif platform.system() == 'Windows' :
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else :
    print ('Unknown system... sorry~!')
    
%matplotlib inline
```

```python
plt.figure(figsize=(12,6))
ko.plot(30)
plt.show()
```

```python
ko.count('고객')
```

```python
plt.figure(figsize=(12,6))
ko.dispersion_plot(['회사', '고객', '컨설턴트'])
```

```python
ko.concordance('회사')
```

```python
ko.collocations()
```

```python
from wordcloud import WordCloud

data = ko.vocab().most_common(150)
wordcloud = WordCloud(font_path='C:\Windows\Fonts\malgun.ttf',
                     relative_scaling=0.2,
                     background_color='white',
                     ).generate_from_frequencies(dict(data))
plt.figure(figsize=(12,8))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
